```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.preprocessing import LabelEncoder

# 1. Load pre-saved data
images = np.load("images.npy", allow_pickle=True)
labels = np.load("labels.npy", allow_pickle=True)

print("Images shape:", images.shape)
print("Total images:", len(images))
print("Unique people:", len(np.unique(labels)))

# 2. Encode labels
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)
print("Encoded label sample:", encoded_labels[:10])

# 3. Dataset summary
counts = Counter(labels)
print("Samples per person:", counts)

# 4. Visualize random images
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    idx = np.random.randint(0, len(images))
    ax.imshow(images[idx], cmap='gray')
    ax.set_title(labels[idx])
    ax.axis('off')
plt.tight_layout()
plt.show()

# 5. Normalize pixel values
images = images / 255.0
print("Data normalized. Min:", np.min(images), "Max:", np.max(images))
```
